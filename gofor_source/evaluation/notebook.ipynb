# Experimentation Notebook for Gofor-SR

In this notebook, you'll find the various experimentations we ran to obtain the results presented in our paper. This notebook calls functions from the '''sim.py''' scripts, which calls the programs found in the '''code''' folder and stores the results in the '''data''' folder in an arborescence generated automatically

**The commands presented here to be run are exactly (and only) the ones used to obtain the figures seen in the paper, although the results may slightly vary due to the randomness of the generated topologies.**

**Experiments simply have to be launched by running each cell in order. Topologies are generated automatically. The script to plot the results are also available, and show the figures directly within the notebook.**

:bulb: Note that the experimentation parameters can however be changed as desired. The accepted parameters are presented below. The interest reader is referred to the cells to find concrete example.

- uses-cases : The use-cases to run. Accepted use-cases are DCLCNoSR, DCLCSR-SRG, DCLCSR-{lex,pareto}-{loose-strict}-{onebest,allbest,all}. Several use-cases can be run at once. In this case, they must be separated by ':'.

- n : The number of nodes. May be a single integer, or a range described as the min, max and step seprated by semicolon (e.g., 100:1000:100)

- mode : The topological model to be used. Accepted arguments are erdos, grid and cross (a grid with diagonal links added). Each model can be pre-pended by "multi" to enabled parallel links. In this case, the propability of appearance of parallel links must be specified. For example, multigrid-0.5 is a grid where half of the links are doubled. 

- degree : Only used by the erdos model to specify the target degree of each node. When not using the erdos model, leave blank. 

- weight : The valuation method to add weight to edge. Accepted parameters are uniform, discrete and equal. Uniform means that values are chosen uniformly at random. In this case, one has to specify the maximum delay a link may have with respect to the constraint. For example, uniform-0.5 means that at most, the delay of a link will be half of the constrained (when considering DCLC). Discrete means that values are taken from a sample of discrete values, and equal means that all valuation are set to the same value. In this case, write equal-0 or discrete-0. 

- force : force the relaunch of the experiments. If not used, only the experiments that were not yet run are launched.
